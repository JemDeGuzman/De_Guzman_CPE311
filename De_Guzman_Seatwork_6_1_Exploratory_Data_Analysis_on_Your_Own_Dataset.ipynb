## Dataset: Netflix Weekly Views Data
### Dataset Link: https://www.kaggle.com/datasets/risakashiwabara/netfllix-weekly-views-data/data



```python
# Importing the Dataset
import numpy as np
import pandas as pd

filepath = "/content/netflixWeeklyViews.csv"
df = pd.read_csv(filepath)
df
```

```python
# Renaming Columns
df.rename(columns = {'week': 'Week', 'category':'Category', 'weekly_rank':'Weekly Rank', 'show_title':'Show Title', 'season_title':'Season Title', 'weekly_hours_viewed':'Weekly View Hours', 'runtime':'Runtime in Hours', 'weekly_views':'Weekly Views', 'cumulative_weeks_in_top_10':'Weeks in Top 10', 'is_staggered_launch':'Rolling Schedule'}, inplace=True)
df
```

```python
# Removing all entries with NaN Weekly Views
df = df[df['Weekly Views'].isna() == False]
df
```

```python
# Creating new DataFrame for English Films
english_films = df[(df['Category']) == 'Films (English)']
english_films
```

```python
# Describing English Films Data
english_films.describe().round(2)
```

```python
# Creating new DataFrame for English TV Shows
english_shows = df[df['Category'] == 'TV (English)']
english_shows
```

```python
# Describing English TV Shows Data
english_shows.describe().round(1)
```

```python
# Top 10 English Films based on Weekly Views
english_films.sort_values('Weekly Views', ascending=False).head(10)
```

```python
# Top 10 English TV Shows based on Weekly Views
english_shows.sort_values('Weekly Views', ascending=False).head(10)
```
