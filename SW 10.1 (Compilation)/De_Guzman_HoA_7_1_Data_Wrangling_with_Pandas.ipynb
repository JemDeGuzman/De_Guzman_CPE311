### Exercise 1

We want to look at data for Facebook, Apple, Amazon, Netflix, and Google (FAANG) stocks, but we were given each as a separate CSV file. Combine them into a single file and store the dataframe of the FAANG data as faang for the rest of the exercises:

1. Read each file in.

```python
import pandas as pd

# Facebook Stocks Dataframe
fb = pd.read_csv('/content/fb.csv')

# Apple Stocks Dataframe
aapl = pd.read_csv('/content/aapl.csv')

# Amazon Stocks Dataframe
amzn = pd.read_csv('/content/amzn.csv')

# Netflix Stocks Dataframe
nflx = pd.read_csv('/content/nflx.csv')

# Google Stocks Dataframe
goog = pd.read_csv('/content/goog.csv')
```

2. Add a column to each dataframe, called **ticker**, indicating the ticker symbol it is for (Apple's is AAPL, for example). This is how you look up a stock. Each file's name is also the ticker symbol, so be sure to capitalize it.

```python
# Adding 'ticker' column to each DataFrame
fb['ticker'] = ["FB" for i in fb.index]
aapl['ticker'] = ["AAPL" for i in aapl.index]
amzn['ticker'] = ["AMZN" for i in amzn.index]
nflx['ticker'] = ["NFLX" for i in nflx.index]
goog['ticker'] = ["GOOG" for i in goog.index]
```

3. Append them together in a single dataframe.

```python
# Concatenating every DataFrame into a single DataFrame
faang = pd.concat([fb, aapl, amzn, nflx, goog])
faang
```

4. Save the result in a CSV file called faang.csv

```python
# Saving the DataFrame as a CSV File
faang.to_csv('/content/faang.csv', index=False)
```

### Exercise 2

*   With faang, use type conversion to change the date column into a datetime and the volume column into integers. Then, sort by date and ticker



```python
# Checking datatypes of columns
faang.dtypes
```

```python
# Converting date column datatype to datetime
faang.loc[:,'date'] = pd.to_datetime(faang.date)
faang.dtypes
```

```python
# Converting volume column datatype to integer
faang.loc[:,'volume'] = pd.to_numeric(faang.volume)
faang.dtypes
```

```python
# Sorting the DataFrame by date and ticker
faang = faang.sort_values(by=['date', 'ticker'], ascending=True)
faang.head(10)
```

*   Find the seven rows with the highest value for volume.

```python
# Finding the 7 rows with highest volume
faang.sort_values(by='volume', ascending=False).head(7)
```

*   Right now, the data is somewhere between long and wide format. Use melt() to make it completely long format. Hint: date and ticker are our ID variables (they uniquely identify each row). We need to melt the rest so that we don't have separate columns for open, high, low, close, and volume.

```python
# Melting the DataFrame into a Long Format DF
faang = faang.melt(
    id_vars = ['date', 'ticker'],
    value_vars = ['open', 'high', 'low', 'close', 'volume'],
)

faang.head(10)
```

### Exercise 3

*   Using web scraping, search for the list of the hospitals, their addresses and contact information. Save the list in a new csv file, hospitals.csv.



```python
from bs4 import BeautifulSoup
import requests
```

```python
# Requesting OutsideHTML of the website
url = 'https://sulit.ph/list-of-hospitals-in-metro-manila-with-contact-details-website-and-social-media-accounts/'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
```

```python
# Finding the table from the HTML
table = soup.find('table')
```

```python
# Finding the table headers
labels = table.find_all('th')
```

```python
# Storing just the header values in a list
table_categories = [title.text for title in labels]
table_categories
```

```python
# Creating a DataFrame using the columns list
hospitals = pd.DataFrame(columns = table_categories)
hospitals
```

```python
# Finding all the row data from the table
column_data = table.find_all('tr')
```

```python
# Adding each row value into the DataFrame
for row in column_data[3:]:
  row_data = row.find_all('td')
  indiv_row_data = [data.text for data in row_data]

  length = len(hospitals)
  hospitals.loc[length] = indiv_row_data

hospitals
```

```python
# Saving the DataFrame as a CSV file
hospitals.to_csv('/content/hospitals.csv', index=False)
```
