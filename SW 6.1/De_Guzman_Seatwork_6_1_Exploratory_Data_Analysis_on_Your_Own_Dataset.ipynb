## Dataset: Netflix Weekly Views Data
### Dataset Link: https://www.kaggle.com/datasets/risakashiwabara/netfllix-weekly-views-data/data



```python
# Importing the Dataset
import numpy as np
import pandas as pd

filepath = "/content/netflixWeeklyViews.csv"
df = pd.read_csv(filepath)
df
```

```python
# Renaming Columns
df.rename(columns = {'week': 'Week', 'category':'Category', 'weekly_rank':'WeeklyRank', 'show_title':'ShowTitle', 'season_title':'SeasonTitle', 'weekly_hours_viewed':'WeeklyViewHours', 'runtime':'RuntimeInHours', 'weekly_views':'WeeklyViews', 'cumulative_weeks_in_top_10':'WeeksInTop10', 'is_staggered_launch':'RollingSchedule'}, inplace=True)
df
```

```python
# Removing all entries with NaN Weekly Views
df = df[df['WeeklyViews'].isna() == False]
df
```

```python
# Creating new DataFrame for English Films
english_films = df[(df['Category']) == 'Films (English)']
english_films
```

```python
# Describing English Films Data
english_films.describe().round(2)
```

```python
# Creating new DataFrame for English TV Shows
english_shows = df[df['Category'] == 'TV (English)']
english_shows
```

```python
# Describing English TV Shows Data
english_shows.describe().round(1)
```

```python
# Top 10 English Films based on Weekly Views
english_films.sort_values('WeeklyViews', ascending=False).tail(10)
```

```python
# Top 10 English TV Shows based on Weekly Views
english_shows.sort_values('WeeklyViews', ascending=False).head(10)
```

```python
# Creating a New Column called 'Watch Time per View'
# It shows how much hours of the media have been
# watched for each individual view it has per week.

df["WatchTimePerView"] = df["WeeklyViewHours"] / df["WeeklyViews"]

watchtime = df[['Week','ShowTitle','WatchTimePerView', 'RuntimeInHours', 'WeeklyViews']]
watchtime_films = watchtime[(df['Category'].str.contains('Films'))]

# Comparing the Top 10 Films (regardless of sub-category) runtime
# and watch time per view
watchtime_films.sort_values('WeeklyViews', ascending=False).head(10)
```

```python
# Comparing the Bottom 10 Films (regardless of sub-category) runtime
# and watch time per view
watchtime_films.sort_values('WeeklyViews', ascending=False).tail(10)
```

```python
# Comparing the Top 10 TV Shows (regardless of sub-category) runtime
# and watch time per view
watchtime_shows = watchtime[(df['Category'].str.contains('TV '))]
watchtime_shows.sort_values('WeeklyViews', ascending=False).head(10)
```

```python
# Comparing the Top 10 TV Shows (regardless of sub-category) runtime
# and watch time per view
watchtime_shows.sort_values('WeeklyViews', ascending=False).tail(10)
```

```python
# Top 10 Films that held Top 1 of the Week longest
rankings = df[['Week', 'Category','ShowTitle', 'WeeklyRank','WeeksInTop10','RollingSchedule']]
firsts = rankings[(rankings['WeeklyRank'] == 1)]
consistent_firsts = firsts[(firsts['Category'].str.contains('Films'))]
consistent_firsts.sort_values('WeeksInTop10', ascending=False).head(10)
```

```python
# Top 10 TV Shows that held Top 1 of the Week longest
consistent_firsts = firsts[(firsts['Category'].str.contains('TV'))]
consistent_firsts.sort_values('WeeksInTop10', ascending=False).head(10)
```
